Monte Carlo simulation. Radial distribution function.

```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

Arrangement of particles on the lattice and RDF for density 0.7.

```python
data07 = pd.read_csv('/Users/bartlomiejkos/Documents/Programming/C++/Wprowadzenie do C++/Symulacj komputerowe w fizyce/RDF_MC/data07.csv', header=None, names=('x','y'))
rdf07 = pd.read_csv('/Users/bartlomiejkos/Documents/Programming/C++/Wprowadzenie do C++/Symulacj komputerowe w fizyce/RDF_MC/RDFdata07.csv', header=None, names=('rw', 'rdf'))
divider = int(len(data07) / 2)
plt.figure(figsize=(20, 10))
plt.subplot(1,2,1)
plt.scatter(data07['x'][:divider:1], data07['y'][:divider:1], s=200, label='initial', c='r', alpha=0.7)
plt.scatter(data07['x'][divider::1], data07['y'][divider::1], s=200, label='final', c='m', alpha=0.7)
plt.xlim(0,10)
plt.ylim(0,10)
plt.xlabel('x [$\sigma$]')
plt.ylabel('y [$\sigma$]')
plt.legend()
plt.title('Arrangement of particles 'r'$\rho$' "= 0.7, T = 0.7K")
plt.subplot(1,2,2)
plt.scatter(rdf07['rw'], rdf07['rdf'], c='r')
plt.plot(rdf07['rw'], rdf07['rdf'], c='m')
plt.ylabel('g(r)')
plt.xlabel(r'$r / \sigma$')
plt.title('RDF for 'r'$\rho$' "= 0.7, T = 0.7K")
plt.grid()
plt.show()
```
